```python
def extract_salary(string):
    
    string = string.split()
    
    return [s for s in string if any([x for x in s if x.isnumeric()])]
```

```python
string = df_new["description"][1]

print(string)
```

```python
extract_salary(string)
```

```python
df_new["prueba_salary"] = df_new["description"].apply(lambda x : extract_salary(x))
```
