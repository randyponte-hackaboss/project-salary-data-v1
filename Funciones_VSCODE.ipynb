# Experience & Experience_Levels

```python
dicc_experience_level = {"junior"      : "Junior",
                         "aprendiz"    : "Junior",
                         "becario"     : "Junior",
                         "jr."         : "Junior",
                         "semisenior"  : "Semi-Senior",
                         "semi-senior" : "Semi-Senior",
                         "semi/senior" : "Semi-Senior",
                         "senior"      : "Senior",
                         "arquitecto"  : "Senior",
                         "coordinador" : "Senior",
                         "leader"      : "Leader",
                         "jefe"        : "Leader"}
```

```python
def find_years_of_experience_title(string):
    
    titulos = ["junior", "aprendiz", "becario", "jr.", "semisenior", "semi-senior", "semi/senior", "senior", "arquitecto",
               "coordinador", "leader"]
    
    for titulo in titulos:
        if titulo in string:
            return titulo
        
    else:
        return np.nan
        
def get_experience_level(string, dicc_experience_level):
    
    return dicc_experience_level[string.lower()].title()
    

def find_years_of_experience_description(string):
    
    list_strings = ["años de", "years of", "years experience", "años experiencia"]
    
    string = string.lower()
    
    years = [string[string.find(s) - 5 : string.find(s) + len(s) + 1] for s in list_strings if string.find(s) != -1]
    
    numeros = [re.findall(r"\d+", y) for y in years]

    numeros = [[int(n) for n in num if 0 < int(n) < 13] for num in numeros]
    
    numeros = [max(num) if num else np.nan for num in numeros]

    return max(numeros) if numeros else np.nan


def experience_level(num):
    
    if not pd.isna(num):
    
        if num < 2:
            return "Junior"
        elif num <= 4:
            return "Semi-Senior"
        elif num < 8:
            return "Senior"
        else:
            return "Leader"
        
    else:
        return np.nan
```

```python
df["experience_levels"] = df["title"].apply(lambda x : find_years_of_experience_title(x.lower()))
```

```python
df["experience"] = df["description"].apply(lambda x : find_years_of_experience_description(x))
```

```python
df["experience_levels"] = [experience_level(y) if str(x) == "nan" else x for x, y in df[["experience_levels", "experience"]].values]
```

```python
dicc_exper_levels = {x : x.lower() for x in df["experience_levels"].value_counts().index}

df["experience_levels"] = df["experience_levels"].map(dicc_exper_levels)

df["experience_levels"] = df["experience_levels"].apply(lambda x : get_experience_level(x, dicc_experience_level) if not pd.isna(x) else x)
```
