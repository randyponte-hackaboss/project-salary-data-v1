```python
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import numpy as np
import json
import pandas as pd
from playsound import playsound #pip install playsound

##Constants
url_infojobs = "https://www.infojobs.net"

##Automation open Chrome in Linux
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

##Automation open Chrome in Windows
#driver = webdriver.Chrome(executable_path = chrome_driver)


#Data from InfoJobs
category_ids = {
    "10"  : "Administración de empresas",
    "20"  : "Administración Pública",
    "30"  : "Calidad, producción e I+D",
    "40"  : "Compras, logística y almacén",
    "50"  : "Diseño y artes gráficas",
    "60"  : "Educación y formación",
    "70"  : "Finanzas y banca",
    "80"  : "Ingenieros y técnicos",
    "90"  : "Inmobiliario y construcción",
    "100" : "Legal",
    "110" : "Marketing y comunicación",
    "120" : "Profesiones, artes y oficios",
    "130" : "Recursos humanos",
    "140" : "Sanidad y salud",
    "150" : "Informática y telecomunicaciones",
    "160" : "Turismo y restauración",
    "170" : "Atención a clientes",
    "180" : "Otros",
    "190" : "Comercial y ventas",
    "200" : "Ventas al detalle",
    "210" : "Sector Farmacéutico"
}

experience_years = {
    "_0_YEARS"  :  0,
    "_1_YEAR"   :  1,
    "_2_YEARS"  :  2, 
    "_3_YEARS"  :  3, 
    "_4_YEARS"  :  4, 
    "_5_YEARS"  :  5, 
    "_10_YEARS" : 10,
                   }
```

```python
def open_browser(category_id, year, page):    
    url_infojobs_jobsearch = f"{url_infojobs}/ofertas-trabajo?keyword=&normalizedJobTitleIds=&provinceIds=&cityIds=&teleworkingIds=&categoryIds={category_id}&workdayIds=&educationIds=&segmentId=&contractTypeIds=&page={page}&sortBy=PUBLICATION_DATE&onlyForeignCountry=false&countryIds=&sinceDate=ANY&experienceMin={year}&experienceMax={year}&subcategoryIds="
    return driver.get(url_infojobs_jobsearch)
```

```python
def get_data(category_value,year_value):
    for x in range(5):
        driver.execute_script("window.scrollBy(0, 1500);")
        time.sleep(3)

    job_title = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-title-link")
    company_name = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-subtitle-link")
    category = category_value
    presence = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-list-item")
    city = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-list-item-truncate")
    contract_type = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-list-item ij-OfferCardContent-description-list-item--hideOnMobile")
    workday = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-list-item ij-OfferCardContent-description-list-item--hideOnMobile")
    salary_range = driver.find_elements(By.CLASS_NAME,"ij-OfferCardContent-description-salary-info")
    experience_min = year_value
    
    lenght = len(job_title)
    print(len(job_title))
       
    
    for e in range(lenght):
        
        job_title_list.append(job_title[e].text)
        
        try:
            company_name_list.append(company_name[e].text)
        except:
            company_name_list.append(np.nan)
            
        try:
            category_list.append(category)
        except:
            category_list.append(np.nan)
            
        try:
            presence_list.append(presence[e].text)
        except:
            presence_list.append(np.nan)
            
        try:
            city_list.append(city[e].text)
        except:
            city_list.append(np.nan)
            
        try:
            contract_type_list.append(contract_type[e].text)
        except:
            contract_type_list.append(np.nan)
            
        try:
            workday_list.append(workday[e].text)
        except:
            workday_list.append(np.nan)
              
        try:
            salary_list.append(salary[e].text)
        except:
            salary_list.append(np.nan)
              
        try:
            salary_range_list.append(salary_range[e].text)
        except:
            salary_range_list.append(np.nan)
            
        experience_min_list.append(experience_min) 
        time.sleep(2) 
```

```python
job_title_list = list()
company_name_list = list()
category_list = list()
presence_list = list()
city_list = list()
contract_type_list = list()
workday_list = list()
salary_list = list()
salary_range_list = list()
experience_min_list = list()
    
for category_id in category_ids.keys():
    category_value = category_ids[category_id]
    
    for year in experience_years.keys():
        year_value = experience_years[year]
        for page in range(100):
            try:
                
                open_browser(category_id, year, str(page+1))
                time.sleep(1)
                driver.maximize_window()
                time.sleep(1)
                for i in range(20):
                    try:
                        captcha = driver.find_element(By.CLASS_NAME,"js-show-captcha-detail").text  
                    except:
                        captcha = False
                    if captcha:
                        playsound("mario-bros_game-over.mp3")                        
                        time.sleep(2)
                    else:
                        pass
                get_data(category_value,year_value)
                
            except:
                pass
            

df_jobs = pd.DataFrame()

df_jobs["job_title"] = job_title_list
df_jobs["company_name"] = company_name_list
df_jobs["category"] = category_list
df_jobs["presence"] = presence_list
df_jobs["city"] = city_list
df_jobs["contract_type"] = contract_type_list
df_jobs["workday"] = workday_list
df_jobs["salary"] = salary_list
df_jobs["salary_range"] = salary_range_list
df_jobs["experience_min"] = experience_min_list

data_json = df_jobs.to_json(orient="records")
load_data = json.loads(data_json)

with open("infojobs_offers_data_10-70.json","w") as file:
    data = json.dumps(load_data,ensure_ascii=False, indent=4)
    file.write(data)
print("Extracción terminada...")
```
