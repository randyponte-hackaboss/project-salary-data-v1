```python
import numpy as np
import pandas as pd

import re

from collections import Counter
from textblob import TextBlob
```

```python
df = pd.read_pickle("prueba_data_science_10pages_es_20230429.pkl")

df
```

```python
def find_skill(word, string):
    
    pattern = r"\b{}\b".format(re.escape(word))
    
    match = re.search(pattern, string)
    
    if match:
        return True
    
    else:
        return False
```

```python
def clean_description(string):
    
    cleaned_string = re.sub(r"\n", " ", string)

    # Matches newline characters or non-alphanumeric characters except $, €, £, and spaces
    pattern = r"[\n]|[^a-zA-Z0-9$€£\s]+"  
    cleaned_string = re.sub(pattern, "", cleaned_string)
    
    return cleaned_string
```

```python
df["description_clean"] = df["description"].apply(lambda x : clean_description(x))
```

### Language Detection

```python
from langdetect import detect

def detect_language(text):
    try:
        language = detect(text)
        return language
    except Exception as e:
        print("Error:", str(e))
        return None

# Example usage
text = "Испания"
language = detect_language(text)
if language:
    print("Detected language:", language)
else:
    print("Language detection failed.")
```
