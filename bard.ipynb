```python
def number_filter(string):
    
    strings = string.split("\n")
    
    strings = [x for x in strings if sum([s.isnumeric() for s in x]) > 3]
    
#     count = [x.isnumeric() for x in string]  
    
    if strings != []:
        
        return True
        
    else:
        return False
```

```python
df1["filter1"] = df1["description"].apply(lambda x : number_filter(x) if not pd.isna(x) else x)
```

```python
df2 = pd.read_csv("LATAM 002-Grid view.csv")
df3 = pd.read_csv("LATAM 003-Grid view.csv")

df2["filter1"] = df2["description"].apply(lambda x : number_filter(x) if not pd.isna(x) else x)
df3["filter1"] = df3["description"].apply(lambda x : number_filter(x) if not pd.isna(x) else x)

resultados1, resultados2, resultados3 = list(), list(), list()
```

```python
%%time

# resultados = list()

for cond, desc, i in zip(df1["filter1"], df1["description"], range(df1.shape[0])):
    
    if not pd.isna(desc):

        if cond:
            print(i)

            desc = " ".join([x for x in desc.split("\n") if sum([y.isnumeric() for y in x]) > 3])

            if desc != "":

                resultados1.append(document_qa(question="Encuentra el salario", context=desc))

        else:
            resultados1.append(np.nan)
            
    else:
        resultados1.append(np.nan)
```
