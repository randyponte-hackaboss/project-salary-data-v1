```python
### Funciones

def url_transform(x):
    
    return x.replace(" ", "-").lower()
```

```python
import requests

wikipedia_provincias = "https://es.wikipedia.org/wiki/Anexo:Provincias_y_ciudades_aut%C3%B3nomas_de_Espa%C3%B1a"

response = requests.get(wikipedia_provincias)

soup_wiki = BeautifulSoup(response.text, "html.parser")

columns = [th.text.strip("\n") for th in soup_wiki.find("tbody").find_all("th")[:-6]]

tabla = [td.text.strip("\n") for td in soup_wiki.find("tbody").find_all("td")]
```

```python
columns
```

```python
tabla = np.array(tabla).reshape(52, 11)


df = pd.DataFrame(tabla, columns = columns)

df.rename(columns = {col : "Superficie_km2" for col in df.columns if col.startswith("Super")}, inplace = True)

df
```

```python
df["Población"] = df["Población"].apply(lambda x : [float(y) for y in x.split("&") if len(y) > 1 and "." in y][0])

df["Superficie_km2"] = df["Superficie_km2"].apply(lambda x : [float(y) for y in x.split("&") if len(y) > 1 and "." in y][0])


df
```
