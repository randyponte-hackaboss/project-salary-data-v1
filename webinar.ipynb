```python
%%time

os.chdir("json_searches/")

dates = list()

df = pd.DataFrame()

for num, file in enumerate(os.listdir()):
    
    print(num)
    
    with open(file = file, mode = "br") as file:
        data = pickle.load(file)
        
    query = data["search_parameters"]["q"]
        
    if ("data" in query) or ("datos" in query):
        if ("SQL" not in query) and ("base" not in query): 
            print(data["search_parameters"]["q"])

            date = datetime.strptime(data["search_metadata"]["created_at"], "%Y-%m-%d %H:%M:%S UTC").date()

            try:

                df_ = pd.json_normalize(data["jobs_results"])

                for i in range(df_.shape[0]):

                    dates.append(date)

                df = pd.concat([df, df_], ignore_index = True)

            except:
                pass


print(df.shape, len(dates))

os.chdir("..")
```

```python
%%time

# Creates "created_date"
df["date_posted"] = dates

# Drops Duplicates
df = df.drop_duplicates("job_id").reset_index(drop = True) 

# Cleans "via"
df["via"] = df["via"].apply(lambda x : clean.clean_source(x))

# Cleans "location"
df["location"] = df["location"].apply(lambda x : clean.clean_location(x))

# Cleans "contract_type"
df["detected_extensions.schedule_type"] = df["detected_extensions.schedule_type"].apply(lambda x : clean.clean_contract_type(x))

# Cleans "created_date"
df["detected_extensions.posted_at"] = df["detected_extensions.posted_at"].apply(lambda x : clean.transform_date(x))

df["date_posted"] = [clean.get_date(x, y) for x, y in df[["date_posted", "detected_extensions.posted_at"]].values]

# Creates "tech_skills"
df["tech_skills"] = df["description"].apply(lambda x : clean.get_skills(x))
```

```python
df = df[df["title"].str.lower().str.contains("science") | df["title"].str.lower().str.contains("scientist") | (df["title"].str.lower().str.contains("científico"))]
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

df_ofertas = pd.DataFrame(df["via"].value_counts().to_dict().items(), columns = ["Portal", "Número Ofertas"]) 

df_ofertas = df_ofertas.sort_values("Número Ofertas").reset_index(drop = True)

plt.barh(df_ofertas["Portal"][-10:], df_ofertas["Número Ofertas"][-10:], color = "green")
plt.title("Portales (for Data Science) - Ofertas Empleo (20/04/2023 - 15/05/2023)", size = 15)
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

df_ofertas = pd.DataFrame(df["location"].value_counts().to_dict().items(), columns = ["Location", "Número Ofertas"]) 

df_ofertas = df_ofertas.sort_values("Número Ofertas").reset_index(drop = True)

df_ofertas.replace("España", "No Especifícado", inplace = True)

plt.barh(df_ofertas["Location"][-10:], df_ofertas["Número Ofertas"][-10:])
plt.title("Ubicación (for Data Science) - Ofertas Empleo (20/04/2023 - 15/05/2023)", size = 15)
plt.grid()
plt.show()
```
